```python
import os

MODEL_FOLDER = 'Results/WildRelationalNetwork_22_17_09'
TRAIN_LOG_FILE = 'train_log.txt'
TRAIN_LOG_FILE_PATH = os.path.join(MODEL_FOLDER, TRAIN_LOG_FILE)
```

```python
import numpy as np

train_acc = []
train_loss = []
val_acc = []

with open(TRAIN_LOG_FILE_PATH, 'r') as f:
    for line in f:
        if 'Epoch' in line:
            line = line.split(':')
            val_acc.append(float(line[2].split(' ')[2]))
            train_acc.append(float(line[3].split(' ')[2]))
            train_loss.append(float(line[4].split(' ')[2]))

x = np.linspace(0, 16, num=len(train_acc))

import matplotlib.pyplot as plt
plt.plot(x, train_acc, label='Train Accuracy')
plt.plot(x, val_acc, label='Validation Accuracy')
#annotate axis
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('WREN model accuracy')
plt.grid(True)
plt.legend()
plt.savefig(os.path.join(MODEL_FOLDER, 'accuracy.png'))
```

```python
plt.plot(x, train_loss, label='train_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('WREN model loss')
plt.grid(True)
plt.legend()
plt.savefig(os.path.join(MODEL_FOLDER, 'loss.png'))
```
